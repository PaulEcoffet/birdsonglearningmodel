```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import birdsonganalysis as bsa
from scipy.io import wavfile
```

# Let see for one syllable

```python
sr, tutor = wavfile.read('../data/ba_syllable_a_end.wav')
sr, synth = wavfile.read('../model/hillclimbing/res/170222_1750/out_0_1.9375746415446167.wav')
sr, ba = wavfile.read('../data/ba_syllable_a_end_out.wav')
```

```python
tfeat = bsa.all_song_features(tutor, sr)
sfeat = bsa.all_song_features(synth, sr)
bfeat = bsa.all_song_features(ba, sr)
```

```python
for fname in tfeat:
    plt.figure(figsize=(10, 5))
    plt.plot(tfeat[fname], label='tutor')
    plt.plot(sfeat[fname], label='synth')
    plt.plot(bfeat[fname], label='ba')
    plt.legend()
    plt.title(fname)
    plt.show()
```

```python
for fname in tfeat:
    print("*"*40)
    print(fname)
    print('tutor: mean: {}, std: {}'.format(np.mean(tfeat[fname]), np.std(tfeat[fname])))
    print('synth: mean: {}, std: {}'.format(np.mean(sfeat[fname]), np.std(sfeat[fname])))
```
